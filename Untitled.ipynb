```python
# https://docs.google.com/spreadsheets/d/1ucH2XBSL_LHUyqxhH4731m6O7vXEDW4X8nFlMd_kO9Q/edit#gid=0
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import rc
%matplotlib inline
from matplotlib import font_manager
f_path = "C:/windows/Fonts/malgun.ttf"
font_manager.FontProperties(fname=f_path).get_name()
rc('font', family='Malgun Gothic')
```

```python
df = pd.read_csv("./data.csv")
```

```python
df_mask = pd.DataFrame(df.groupby("date").sum()).reset_index()
df_mask["ratio"] = df_mask["no_mask"] / df_mask["all"]
df_mask["date"] = pd.to_datetime(df_mask.date)
df_mask = df_mask.sort_values(by="date").reset_index(drop=True)

hour_range = []
for t in df_mask.date:
    if t.hour < 13:
        hour_range.append(0)
    elif t.hour >= 13 and t.hour < 20:
        hour_range.append(1)
    else:
        hour_range.append(2)
        
df_mask["hour_range"] = hour_range
df_mask["hour"] = df_mask.date.apply(lambda x: x.hour)
df_mask.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_mask.date, df_mask.ratio, 'd')
plt.xticks(rotation=5)
plt.title("노마스크 비율")
plt.tight_layout()
# plt.savefig("nomask_ratio.png", facecolor="w")
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df_mask.date, df_mask.ratio, c=df_mask.hour_range)
plt.xticks(rotation=5)
plt.title("노마스크 비율")
plt.tight_layout()
```

```python
plt.figure(figsize=(12, 6))
sc = plt.scatter(df_mask.date, df_mask.ratio, c=df_mask.hour)
plt.xticks(rotation=10, fontsize=12)
plt.yticks(fontsize=12)
plt.title("노마스크 비율, cbar: hour", fontsize=12)
plt.colorbar(sc)
# plt.tight_layout()
plt.savefig("nomask_ratio.png", facecolor="w")
```
